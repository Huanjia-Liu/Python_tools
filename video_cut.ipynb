```python
from moviepy.editor import *
```

```python
def video_clip(start,end,size,input,output,if_logger=False):
    clip = VideoFileClip(input)
    clip = clip.subclip(t_start=start,t_end= end)
    clip = clip.resize(size)
    clip.write_videofile(output)
    clip.close()
```

```python
start = (1,17)
end = (1,24)
size = 0.8
file_input = "E://Video/2021-04-24 23-18-52.mkv"
file_output = "E://Video/0424.mp4"
video_clip(start,end,size,file_input,file_output)
```
